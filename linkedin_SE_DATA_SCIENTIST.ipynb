# Importing Libraries

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import csv
import time
from selenium.webdriver.support import expected_conditions as EC
```

# Creating Required Functions

```python
def check(): #Checking Scroll Exists 
    try:
        WebDriverWait(driver,0.5).until(EC.visibility_of_element_located((By.XPATH, "//*[@aria-label='Load more results']")))
        return False
    except:
        return True
    
    
def extract_Data(driver):
    wait = WebDriverWait(driver, 5)
    location = wait.until(EC.visibility_of_element_located((By.XPATH,"//span[@class='topcard__flavor topcard__flavor--bullet']"))).text
    title = wait.until(EC.visibility_of_element_located((By.XPATH,"//h2"))).text
    date_posted = wait.until(EC.visibility_of_element_located((By.XPATH,"//span[@class='posted-time-ago__text topcard__flavor--metadata']"))).text
    company = wait.until(EC.visibility_of_element_located((By.XPATH,"//span[@class='topcard__flavor']"))).text
    Employement_type = wait.until(EC.visibility_of_element_located((By.XPATH,"//h3[contains(text(),'Employment type')]/../span"))).text
    
    try:
        showMore = WebDriverWait(driver, 2).until(EC.visibility_of_element_located((By.XPATH,"//button[@data-tracking-control-name='public_jobs_show-more-html-btn']")))
        showMore.click()
        
    except:
        pass
    
    details = wait.until(EC.visibility_of_element_located((By.XPATH,"//*[@class='description__text description__text--rich']")))  
    
      # Responsibility

    try:
        responsibility = details.find_elements(By.XPATH, "//strong[contains(text(), 'Responsibilities')]/following-sibling::*")

        if len(responsibility) == 0:
            responsibility = details.find_elements(By.XPATH, "//strong/u[contains(text(), 'Responsibilities')]/../following-sibling::*")
            responsibility = "".join([i.text for i in responsibility])

        else:
            responsibility = "".join([i.text for i in responsibility])

    except Exception as e:
        responsibility = ""


    # Qualifications      

    try:
        qualitfication = details.find_elements(By.XPATH, "//strong[contains(text(), 'Qualifications')]/following-sibling::*")

        if len(qualitfication) == 0:
            qualitfication = details.find_elements(By.XPATH, "//strong/u[contains(text(), 'Qualifications')]/../following-sibling::*")
            qualitfication = "".join([i.text for i in qualitfication])

        else:
            qualitfication = "".join([i.text for i in qualitfication])

    except Exception as e:
        qualitfication = ""

    # Skills

    try:
        skills = details.find_elements(By.XPATH, "//strong[contains(text(), 'Skills')]/following-sibling::*")

        if len(skills) == 0:
            skills = details.find_elements(By.XPATH, "//strong/u[contains(text(), 'Skills')]/../following-sibling::*")
            skills = "".join([i.text for i in skills])

        else:
            skills = "".join([i.text for i in skills])

    except Exception as e:
        skills = ""

    # Requirements

    try:
        requirements = details.find_elements(By.XPATH, "//strong[contains(text(), 'Requirements')]/following-sibling::*")

        if len(requirements) == 0:
            requirements = details.find_elements(By.XPATH, "//strong/u[contains(text(), 'Requirements')]/../following-sibling::*")
            requirements = "".join([i.text for i in requirements])

        else:
            requirements = "".join([i.text for i in requirements])

    except Exception as e:
        requirements = ""

    # Job Description

    try:
        job_desc = details.find_elements(By.XPATH, "//strong[contains(text(), 'Job Description')]/following-sibling::*")

        if len(job_desc) == 0:
            job_desc = details.find_elements(By.XPATH, "//strong/u[contains(text(), 'Job Description')]/../following-sibling::*")
            job_desc = "".join([i.text for i in job_desc])

        else:
            job_desc = "".join([i.text for i in job_desc])

    except Exception as e:
        job_desc = ""


    # Experience


    try:
        experience = details.find_elements(By.XPATH, "//strong[contains(text(), 'Experience')]/following-sibling::*")

        if len(experience) == 0:
            experience = details.find_elements(By.XPATH, "//strong/u[contains(text(), 'Experience')]/../following-sibling::*")
            experience = "".join([i.text for i in experience])

        else:
            experience = "".join([i.text for i in experience])

    except Exception as e:
        experience = ""          
    
    print("Title : ", title)
    print("Date Posted : ", date_posted)
    print("Location : ", location)
    print("Company : ", company)
    print("Details : ", details.text)
    print("Employement Type : ", Employement_type)
    print("Responsibility : ", responsibility)
    print("Qualification : ", qualitfication)
    print("Skills : ", skills)
    print("Requirements : ", requirements)
    print("Experience : ", experience)
    
    data = [title, date_posted, location, company, details.text, Employement_type, responsibility, qualitfication, skills, requirements, experience]
    return data
    
def checkScroll(driver):
    pageHeight = driver.execute_script("return document.body.scrollHeight")
    totalScrolledHeight = driver.execute_script("return window.pageYOffset + window.innerHeight")

    if((pageHeight-1)<=totalScrolledHeight):
        return True
    else:
        return False
```

# Creating and Writing Header to CSV file

```python
with open('Scientist.csv','w', newline='',encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Title", "Date Posted", "Location", "Company", "Job description", "Employement Type", "Responsibility", "Qualification", "Skills", "Requirements", "Experience"])
```

# Chrome Options

```python
co = webdriver.ChromeOptions()
co.add_argument("--start-maximized")
co.add_argument('--disable-gpu')
co.add_argument('--disable-infobars')
```

# Initializing Chromedriver

```python
driver = webdriver.Chrome(ChromeDriverManager().install(),options=co)
```

# Getting Link

```python
driver.get("https://www.linkedin.com/jobs/search?keywords=Scientist&location=Toronto%2C%20Ontario%2C%20Canada&geoId=100025096&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0")
```

# Scrolling till the end of page to get all the available jobs

```python
flag = True
list_ = []
scrollHeight = 1000
while flag:
    driver.execute_script(f"window.scrollTo(0, {scrollHeight});")
    time.sleep(2)

    try:
        element = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, "//*[@aria-label='Load more results']")))
        driver.execute_script("arguments[0].click();", element)
    
    except Exception as e:
        print(e)
        pass
    
    scrollHeight = scrollHeight + 800
    
    if checkScroll(driver) == True:
        list_.append(1)
        
#         Y = driver.execute_script("return document.body.scrollHeight")
#         Y = Y - 300
#         driver.execute_script(f"window.scrollTo(document.body.scrollHeight, {Y});"
        
        if len(list_) == 4:
            flag = False
```

# Scraping All the jobs

```python
jobs_card = driver.find_elements(By.XPATH,"//a[@data-tracking-control-name='public_jobs_jserp-result_search-card']/../..")
print(f"{len(jobs_card)} Jobs Found ")

for count, job in enumerate(jobs_card):
    try:
        driver.execute_script("arguments[0].scrollIntoView();", job)
        time.sleep(2)
        job.click()
        
        print(f"------------------------Job Count is {count + 1}------------------------")
        data = extract_Data(driver)
        
        with open('Scientist.csv','a+', newline='',encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow(data)
        
        time.sleep(3)
        
    except Exception as e:
        print(e)
        pass
    
    
```
